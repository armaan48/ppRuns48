```python
from numpy import NAN
import pandas as pd
import sys
from sklearn import linear_model
import joblib



stadiums = [ 
    "Narendra Modi Stadium",
    "MA Chidambaram Stadium",
    "Arun Jaitley Stadium",
    "M.Chinnaswamy Stadium",
    "Wankhede Stadium",
    "Dr DY Patil Sports Academy",
    "Punjab Cricket Association IS Bindra Stadium",
    "Sawai Mansingh Stadium",
    "Rajiv Gandhi International Stadium",
    "Eden Gardens",
    "Maharashtra Cricket Association Stadium",
    "Brabourne Stadium",
    "Himachal Pradesh Cricket Association Stadium",
    "Bharat Ratna Shri Atal Bihari Vajpayee Ekana Cricket Stadium"
]


team = [
    "Rajasthan Royals",
    "Gujarat Titans",
    "Royal Challengers Bangalore",
    "Lucknow Super Giants",
    "Sunrisers Hyderabad",
    "Punjab Kings",
    "Delhi Capitals",
    "Mumbai Indians",
    "Chennai Super Kings",
    "Kolkata Knight Riders"
]


teamD = {
     "Rajasthan Royals" : 1,
    "Gujarat Titans" : 2 ,
    "Royal Challengers Bangalore" : 3,
    "Lucknow Super Giants" : 4,
    "Sunrisers Hyderabad" : 5,
    "Punjab Kings" : 6,
    "Delhi Capitals" : 7,
    "Mumbai Indians" : 8,
    "Chennai Super Kings" : 9,
    "Kolkata Knight Riders" : 10
}

stadiumD = {
    "Narendra Modi Stadium" : 1,
    "MA Chidambaram Stadium" : 2,
    "Arun Jaitley Stadium" : 3,
    "M.Chinnaswamy Stadium" : 4,
    "Wankhede Stadium" : 5,
    "Dr DY Patil Sports Academy" : 6,
    "Punjab Cricket Association IS Bindra Stadium" : 7,
    "Sawai Mansingh Stadium" : 8,
    "Rajiv Gandhi International Stadium" : 9,
    "Eden Gardens" : 10,
    "Maharashtra Cricket Association Stadium" : 11,
    "Brabourne Stadium" : 12,
    "Himachal Pradesh Cricket Association Stadium" : 13,
    "Bharat Ratna Shri Atal Bihari Vajpayee Ekana Cricket Stadium" : 14
}



df = pd.read_csv("/content/drive/MyDrive/IPLcontest/pastData/ballByball.csv")
fd = pd.read_csv("/content/drive/MyDrive/IPLcontest/pastData/matchResult.csv")
f = dict(fd)
d = dict(df)
n = len(d['ID'])
m = len(f['ID'])
def whichStadium(id):
    for i  in range(m):
        if (f['ID'][i]==id):
            return f['Venue'][i]
    
def score(id , k):
    runs = 0
    wkt = 0
    t = "no"
    for i in range(n):
        if (d['ID'][i]==id and d['innings'][i]==k):
            t = d['BattingTeam'][i]
            wkt += d['isWicketDelivery'][i]
            runs += int(d['total_run'][i])
    
    return [t , runs , wkt]

ST = pd.DataFrame(columns = ['Venue' , 'Wkt' , 'Runs' , 'frq' ])
for st in stadiums:
    for ww in range(11):
        ST.loc[len(ST.index)] = [st, ww , 0 , 0 ]

for name in range( 10  ):
    T1 = pd.DataFrame(columns = ['Opponent' , 'Wkt' , 'Runs' , 'frq'])
    T2 = pd.DataFrame(columns = ['Opponent' , 'Wkt' , 'Runs' , 'frq' ])
    TS = pd.DataFrame(columns = ['Venue' , 'Wkt' , 'Runs' , 'frq'])
    TEAMNAME =  team[name]
    done = []
    for tt in team:
        for ww in range(11):
            T1.loc[len(T1.index)] = [tt, ww , 0 , 0 ]
            T2.loc[len(T2.index)] = [tt, ww , 0 , 0 ]

    for st in stadiums:
        for ww in range(11):
            TS.loc[len(TS.index)] = [st, ww , 0 , 0 ]


    
    
    
    for i in range(n):
        matchID = d['ID'][i]
        stad = whichStadium(matchID)
        nstad = str();
        if (stad==None):
            continue
        for v in stad:
            if (v==','):
                break;
            nstad += v
        stad = nstad
        if (stadiums.count(stad)==0):
            continue
        if done.count(matchID)==0:
            done.append(matchID)
        
            t1 , sc1 , wk1  = score(matchID , 1)
            t2 , sc2 , wk2 = score(matchID , 2)
            

            if (t2==TEAMNAME):
                t1 , t2 =  t2 , t1 
                sc1 , sc2  = sc2 , sc1 
                wk1 , wk2  = wk2,  wk1

            if (t1==TEAMNAME and t2!="no"):
                p1 = T1['Runs'][((T1.Opponent==t2) & (T1.Wkt==wk1) )]
                p2 = T1['frq'][((T1.Opponent==t2) & (T1.Wkt==wk1) )]

                T1.loc[((T1.Opponent==t2)  & (T1.Wkt==wk1) ),'Runs'] = p1 +   sc1
                T1.loc[((T1.Opponent==t2)  & (T1.Wkt==wk1) ),'frq'] = p2 + 1



                p1 = T2['Runs'][((T2.Opponent==t2) & (T2.Wkt==wk2) )]
                p2 = T2['frq'][((T2.Opponent==t2) & (T2.Wkt==wk2) )]

                T2.loc[((T2.Opponent==t2) &  (T2.Wkt==wk2) ),'Runs'] = p1 +   sc2
                T2.loc[((T2.Opponent==t2) &  (T2.Wkt==wk2) ),'frq'] = p2 + 1


                p1 = TS['Runs'][((TS.Venue==stad) & (TS.Wkt==wk1) )]
                p2 = TS['frq'][((TS.Venue==stad) & (TS.Wkt==wk1) )]

                TS.loc[((TS.Venue==stad) &  (TS.Wkt==wk1) ),'Runs'] = p1 +   sc1
                TS.loc[((TS.Venue==stad) &  (TS.Wkt==wk1) ),'frq'] = p2 + 1



                p1 = TS['Runs'][((TS.Venue==stad) & (TS.Wkt==wk1) )]
                p2 = TS['frq'][((TS.Venue==stad) & (TS.Wkt==wk1) )]

                TS.loc[((TS.Venue==stad) &  (TS.Wkt==wk1) ),'Runs'] = p1 +   sc1
                TS.loc[((TS.Venue==stad) &  (TS.Wkt==wk1) ),'frq'] = p2 + 1



                p1 = ST['Runs'][((TS.Venue==stad) & (ST.Wkt==wk1) )]
                p2 = ST['frq'][((TS.Venue==stad) & (ST.Wkt==wk1) )]

                ST.loc[((TS.Venue==stad) &  (ST.Wkt==wk1) ),'Runs'] = p1 +   sc1
                ST.loc[((TS.Venue==stad) &  (ST.Wkt==wk1) ),'frq'] = p2 + 1

    
    
    cnt = 0


    for tt in team:
        for ww in range(11):
            L = T1.loc[cnt]
            if (L[3]):
                T1.loc[cnt] = [tt, ww , L[2]/L[3] , L[3] ]
            L = T2.loc[cnt]
            if (L[3]):
                T2.loc[cnt] = [tt, ww , L[2]/L[3] , L[3] ]

            cnt += 1

    cnt = 0
    for st in stadiums:
        for ww in range(11):
            L = TS.loc[cnt]
            if (L[3]):
                TS.loc[cnt] = [st , ww , L[2]/L[3] , L[3] ]

            cnt += 1
    




    T1['Opponent'] = T1['Opponent'].apply(lambda x: teamD[x])
    T2['Opponent'] = T2['Opponent'].apply(lambda x: teamD[x])
    TS['Venue'] = TS['Venue'].apply(lambda x: stadiumD[x])


    T1 = T1[T1.Runs > 0]
    T2 = T2[T2.Runs > 0]
    TS = TS[TS.Runs > 0]

    reg = linear_model.LinearRegression()
    reg.fit(T1[ [  'Opponent' , 'Wkt'  ] ] , T1.Runs)  
    F  = "P" + str(name) + "1"
    joblib.dump(reg , "/content/drive/MyDrive/IPLcontest/" + F)



    reg = linear_model.LinearRegression()
    reg.fit(T2[ [  'Opponent' , 'Wkt' ] ] , T2.Runs)  
    F  = "P" + str(name) + "2"
    joblib.dump(reg , "/content/drive/MyDrive/IPLcontest/" + F)


    reg = linear_model.LinearRegression()
    reg.fit(TS[ [  'Venue' , 'Wkt'  ] ] , TS.Runs)  
    F  = "P" + str(name) + "S"
    joblib.dump(reg , "/content/drive/MyDrive/IPLcontest/" + F)


cnt = 0
for st in stadiums:
    for ww in range(11):
        L = ST.loc[cnt]
        if (L[3]):
            ST.loc[cnt] = [st , ww , L[2]/L[3] , L[3] ]
        cnt += 1

ST['Venue'] = ST['Venue'].apply(lambda x: stadiumD[x])
ST = ST[ST.Runs > 0]

reg = linear_model.LinearRegression()
reg.fit(ST[ [  'Venue' , 'Wkt'  ] ] , ST.Runs)  
F  = "PSF"
joblib.dump(reg , "/content/drive/MyDrive/IPLcontest/" + F)

```
